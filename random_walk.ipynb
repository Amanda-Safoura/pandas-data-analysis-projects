## Exercice 2 : Random Walk

```python
import numpy as np
import matplotlib.pyplot as plt

all_walks = list()

np.random.seed(123)

for i in range(1000): # 1000 simulations
    
    random_walk = [0]
    for x in range(50): # soit 50 lancers
        step = random_walk[-1]
        essai = np.random.randint(0,2)
    
        if essai == 1:
            step += 1
        else:
            step = max(step-1, 0)
    
        random_walk.append(step)
    
    #print(random_walk)
    all_walks.append(random_walk)

np_walks = np.array(all_walks)
#print(np_walks.shape)
print('Moyenne finale: ' + str(np.mean(np_walks)))
print('Maximum atteint: ' + str(np.max(np_walks)))

last_steps = np_walks[:, -1]
print("Probabilité d'atteindre la 20ème marche: " + str(len(last_steps[last_steps >=20])/1000))
print()

plt.clf()
plt.figure()
#visualisation de la première marche aléatoire
plt.plot(np_walks[0])
plt.show()

plt.clf()
#visualisation des 5 premières marches aléatoires
plt.plot(np.transpose(np_walks[:5]))
plt.show()

plt.clf()
#histogrammes des positions finales
plt.hist(last_steps)
plt.show()
```
